```python
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng()
```

```python
# 1
N = 5
x = rng.random(N)
avg = np.average(x)
dev = np.sqrt(np.mean((x-avg)**2))
mean_dev=dev/np.sqrt(N)

print('Average X:',avg)
print('Standard Dev: ',dev)
print('Standard Dev of Mean: ',mean_dev)
```

```python
Nval = np.logspace(1, 10, num=10, dtype=int)
Nval = np.clip(Nval,10,None)
avg = []
dev = []
mean_dev = []
Na = []

for N in Nval:
    x=rng.random(N)
    avg1 = np.average(x)
    dev1 = np.sqrt(np.mean((x-avg1)**2))
    mean_dev1 = dev / np.sqrt(N)
    avg.append(avg1)
    dev.append(dev1)
    mean_dev.append(mean_dev1)
    Na.append(N)
    
print('N values of: ',Na)
print('Average X:',avg,'\n')
print('Standard Dev: ',dev,'\n')
print('Standard Dev of Mean: ',mean_dev)
```

```python
#As N increase, the average stays around the same values of ~0.5
```

```python
#The Standard Deviation of the Mean gets smaller 
```

```python
#If you are concerned with the measurement of error, Standard Deviation of the Mean is more important with larger sample sizes.
#Where Standard Deviation is more inportant point to point, Standard Deviation of the Mean gives insight into the populations error
```

```python
#2
import re
filename='Flaxseed.csv'
filename1='Apples.csv'
filename2='Rice.csv'
```

```python
yearF = []
bpaF = []
yearA = []
bpaA = []
yearR = []
bpaR = []

csv_split_pattern = r',(?=(?:[^"]*"[^"]*")*[^"]*$)'

with open(filename, "r") as datafile:
    for line in datafile:
        fields = re.split(csv_split_pattern,line.strip())
        fields = [field.strip('"') for field in fields]
        yearF.append(fields[1])  # Year is at index 3
        bpaF.append(fields[18])  # Bu per acre is at index 16

with open(filename1, "r") as datafile1:
    for line in datafile1:
        fields = re.split(csv_split_pattern,line.strip())
        fields = [field.strip('"') for field in fields]
        yearA.append(fields[1])  # Year is at index 3
        bpaA.append(fields[18])  # Bu per acre is at index 16

with open(filename2, "r") as datafile2:
    for line in datafile2:
        fields = re.split(csv_split_pattern,line.strip())
        fields = [field.strip('"') for field in fields]
        yearR.append(fields[1]) # Year is at index 3
        bpaR.append(fields[18])  # Bu per acre is at index 16

print("Flaxseed Years:", yearF)
print("Flaxseed Bu per acre:", bpaF)
print("Apples Years:", yearA)
print("Apples Bu per acre:", bpaA)
print("Rice Years:", yearR)
print("Rice Bu per acre:", bpaR)
```

```python
yearF = [int(year) for year in yearF[1:]]
yearA = [int(year) for year in yearA[1:]]
yearR = [int(year) for year in yearR[1:]]
bpaF = [int(value.replace(',', '')) for value in bpaF[1:]]
bpaA = [int(value.replace(',', '')) for value in bpaA[1:]]
bpaR = [int(value.replace(',', '')) for value in bpaR[1:]]
```

```python
plt.plot(yearF, bpaF, "o", markerfacecolor="none", label="Flaxseed (BU/ACRE)")
plt.plot(yearA, bpaA, "o", markerfacecolor="none", label="Apples (BU/ACRE)")
plt.plot(yearR, bpaR, "o", markerfacecolor="none", label="Rice (BU/ACRE)")

plt.xlabel("Time (Calendar Year)")
plt.ylabel("Average Per Acre Production")
plt.title("US Agriculture Yields, NASS (USDA)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#4
N = 1000
def estimate_pi(N):
    x = rng.uniform(-1,1,N)
    y = rng.uniform(-1,1,N)
    dis = x**2 + y**2
    c = np.sum(dis <= 1)
    Epi = 4 * c / N
    return Epi

target = 1e-6

while True:
    estimate = estimate_pi(N)
    error = np.abs(estimate - np.pi)
    
    if error < target:
        break
    N *= 2

print(f'Estimated Pi: {estimate}')
print(f'Number of samples used: {N}')
print(f'Error: {error}')
```

```python
#5
T = 0.1 
w = 2 * np.pi / T
Io = 2 #Guess?
rng = np.random.default_rng()

def rms(b, Io, num_samples=10000):
    t = rng.uniform(0, T, num_samples)
    I = np.zeros(num_samples)
    I[t <= b] = Io * np.cos(w * t[t <= b])
    Isqr = I**2
    Iavg = np.average(Isqr)
    rms_value = np.sqrt(Iavg)  
    return rms_value

b_values = np.arange(0.001, 0.101, 0.001) 
RMS = []

for b in b_values:
    RMS.append(rms(b, Io))

plt.plot(b_values * 1000, RMS)  # Scale b_values by 1000 to show in ms
plt.xlabel('b (ms)')
plt.ylabel('RMS Current')
plt.title(f'RMS Current as a Function of b (Peak Current = {Io} A)')
plt.grid(True)
plt.show()
```
